```python
# PASO 1: IMPORTAR Y CARGAR DATOS

import os
import geopandas as gpd
import pandas as pd

# 1. Definición de rutas
ruta_entrada = os.path.join('../../datos/crudos/zonas', '2020_Neighborhood_Tabulation_Areas_(NTAs)_20260118.geojson')

# 2. Carga del dataset geoespacial
gdf_ntas = gpd.read_file(ruta_entrada)


print(f"📊 Registros totales (NYC): {len(gdf_ntas)}")
print(f"🌍 Columnas detectadas: {list(gdf_ntas.columns)}")
display(gdf_ntas.head(5))
```

```python
# PASO 2: SELECCIÓN Y LIMPIEZA DE COLUMNAS

# Filtramos por boroname o borocode (Manhattan = 1)
gdf_manhattan = gdf_ntas[
    (gdf_ntas['boroname'] == 'Manhattan') | (gdf_ntas['borocode'] == '1')
].copy()

# Seleccionamos las columnas que realmente aportan valor al análisis espacial
columnas_interes = [
    'borocode', 'boroname', 'nta2020', 'ntaname', 
    'cdta2020', 'cdtaname', 'shape_leng', 'shape_area', 'geometry'
]

gdf_manhattan = gdf_manhattan[columnas_interes]

# Renombrado de variables (menos geometry)
mapeo_columnas = {
    'borocode': 'ID_DISTRITO',
    'boroname': 'NOMBRE_DISTRITO',
    'nta2020': 'ID_ZONA',
    'ntaname': 'NOMBRE_ZONA',
    'cdta2020': 'ID_CDTA',
    'cdtaname': 'NOMBRE_CDTA',
    'shape_area': 'AREA_M2',
    'shape_leng': 'PERIMETRO_M',
}
gdf_manhattan = gdf_manhattan.rename(columns=mapeo_columnas)


print(f"📍 Barrios en Manhattan: {len(gdf_manhattan)}")
print("-" * 70)
display(gdf_manhattan.head(5))
```

```python
# PASO 3: ESTANDARIZACIÓN Y CONVERSIÓN DE TIPOS DE DATOS (GEOJSON)

# Inspección del tipo de datos
print("\n📊 Resumen de la estructura:")
gdf_manhattan.info()

# Conversión de tipos numéricos y decimales
gdf_manhattan['ID_DISTRITO'] = pd.to_numeric(gdf_manhattan['ID_DISTRITO'], errors='coerce').astype(int)
gdf_manhattan['AREA_M2'] = gdf_manhattan['AREA_M2'].astype(float)
gdf_manhattan['PERIMETRO_M'] = gdf_manhattan['PERIMETRO_M'].astype(float)

# Auditoría final de tipos
print("\n📊 TIPOS DE DATOS FINALES:")
print(gdf_manhattan.dtypes)
```

```python
# PASO 4: AUDITORÍA DE CALIDAD DE LOS DATOS (MTA)

print("🔍 AUDITORÍA DE CALIDAD")
print("-" * 70)

# Verificamos nulos en las nuevas columnas
nulos = gdf_manhattan.isnull().sum()
print(f"📝 Nulos detectados:\n{nulos}")

# Verificamos que la columna GEOMETRIA sea válida
invalidos = (~gdf_manhattan.geometry.is_valid).sum()
print(f"⚠️ Geometrías inválidas: {invalidos}")
```

```python
# PASO 5: EXPORTACIÓN FINAL DEL GEOJSON

import os

# Definición del nombre del archivo y ruta para guardar 
carpeta = '../../datos/limpios/zonas'
nombre_archivo = '12_NTAs_Manhattan_Geometry.geojson'
ruta_completa = os.path.join(carpeta, nombre_archivo)


try:
    # Guardamos el archivo en formato GeoJSON
    gdf_manhattan.to_file(ruta_completa, driver='GeoJSON')
    
    print(f"📂 Archivo 12 guardado en: {ruta_completa}")
    print(f"📊 Barrios exportados: {len(gdf_manhattan)}")
    print(f"🌍 Sistema de Coordenadas: {gdf_manhattan.crs}")

except Exception as e:
    print(f"❌ Error al guardar el archivo: {e}")
```
