```python
# PASO 1: Carga y Preparación del Dataset Maestro

import pandas as pd
import numpy as np
import os

# Configuración de rutas del repositorio
archivo_fuente = "../../datos/maestro/MASTER_DATASET_MANHATTAN_ML.csv"

print("-" * 70)
print("🚀 PASO 1: CARGA Y PREPARACIÓN DEL DATASET MAESTRO")
print("-" * 70)

# Carga del archivo
if os.path.exists(archivo_fuente):
    df_maestro = pd.read_csv(archivo_fuente)

    print(f"📊 Dimensiones: {df_maestro.shape[0]} filas y {df_maestro.shape[1]} columnas.")
    print("-" * 70)
    display(df_maestro.head())
else:
    print(f"❌ ERROR: No se encontró el archivo en {archivo_fuente}")

print("-" * 70)
print("✅ PASO 1 COMPLETADO: Dataset examinado y listo.")

```

```python
# PASO 2: CÁLCULO DE PERCENTILES (P5 Y P95)

print("-" * 70)
print("🚀 PASO 2: CÁLCULO DE PERCENTILES (P5 Y P95)")
print("-" * 70)

# 2.1 Definición de las 16 variables analíticas según su sentido estratégico
variables_directas = [
    'POBLACION_KM2', 'PORCENTAJE_HISPANOS', 'EDAD_MEDIANA', 
    'INGRESO_MEDIANO_HOGAR', 'TAMANO_HOGAR_PROMEDIO',
    'MOVILIDAD_PROMEDIO_DIARIA', 'MOV_CANTIDAD_ESTACIONES',
    'LUGARES_COMERCIO_KM2', 'LUGARES_OFICINAS_KM2', 
    'LUGARES_RESIDENCIAL_KM2', 'COMPETENCIA_INDIRECTA_KM2'
]

variables_inversas = [
    'DELITO_PROPIEDAD_KM2', 'DELITO_TRANSPORTE_KM2', 'DELITO_OTROS_KM2',
    'COMPETENCIA_DIRECTA_KM2', 'ALQ_PRECIO_PIE2_ANUAL'
]

todas_variables = variables_directas + variables_inversas

# 2.2 Diccionario para almacenar los límites
limites_percentiles = {}

# 2.3 Cálculo de límites para mitigar outliers
for var in todas_variables:
    p5 = df_maestro[var].quantile(0.05)
    p95 = df_maestro[var].quantile(0.95)
    limites_percentiles[var] = {'p5': p5, 'p95': p95} 

# 2.4 Visualización de los límites establecidos
print("📋 LÍMITES DE NORMALIZACIÓN (P5 - P95):")
print("-" * 60)
for var, val in limites_percentiles.items():
    print(f"🔹 {var:25} | P5: {val['p5']:10.2f} | P95: {val['p95']:10.2f}")
```

```python
# PASO 3: NORMALIZACIÓN CON SENTIDO ESTRATÉGICO (0-100)

print("-" * 70)
print("🚀 PASO 3: NORMALIZACIÓN CON SENTIDO ESTRATÉGICO (0-100)")
print("-" * 70)

# Copia del dataset original
df_scoring = df_maestro.copy()

# 3.1 Aplicación de fórmulas de normalización
for var in todas_variables:
    p5 = limites_percentiles[var]['p5']
    p95 = limites_percentiles[var]['p95']
    
    # Aplicamos el "clipeo" para obligar a los valores a estar entre P5 y P95
    df_scoring[var] = df_scoring[var].clip(lower=p5, upper=p95)
    
    # Normalización según el sentido estratégico
    if var in variables_directas:
        # Sentido Directo: (Valor - P5) / (P95 - P5) * 100
        df_scoring[f'SCORE_{var}'] = ((df_scoring[var] - p5) / (p95 - p5)) * 100
    else:
        # Sentido Inverso: (P95 - Valor) / (P95 - P5) * 100
        df_scoring[f'SCORE_{var}'] = ((p95 - df_scoring[var]) / (p95 - p5)) * 100

# 3.2 Validación de rangos [0-100]
columnas_score = [f'SCORE_{v}' for v in todas_variables]
resumen_scores = df_scoring[columnas_score].agg(['min', 'max']).T

print("📊 VALIDACIÓN DE RANGOS DE PUNTUACIÓN (0 - 100):")
print("-" * 60)
display(resumen_scores)
```

```python
# PASO 4: INCLUSIÓN DE ETIQUETA DE CLUSTER

print("-" * 70)
print("🚀 PASO 4: INCLUSIÓN DE ETIQUETA DE CLUSTER")
print("-" * 70)

# 4.1 Carga del archivo de resultados de clustering seleccionado (Ward k=4)
archivo_clusters = "../../resultados/04_clustering/Ward.D/manhattan_Ward_k4.xlsx"

if os.path.exists(archivo_clusters):
    df_clusters_ext = pd.read_excel(archivo_clusters)
    df_clusters_ext = df_clusters_ext[['ID_ZONA', 'Cluster']]

    if 'CLUSTER' in df_scoring.columns:
        df_scoring.drop(columns=['CLUSTER'], inplace=True)

    df_scoring = pd.merge(df_scoring, df_clusters_ext, on='ID_ZONA', how='left')
    df_scoring.rename(columns={'Cluster': 'CLUSTER'}, inplace=True)

    print(f"✅ Etiqueta de Cluster integrada desde: {archivo_clusters}")
    print("📊 Distribución de zonas por Cluster:")
    print(df_scoring['CLUSTER'].value_counts().sort_index())
    print("-" * 70)

    pd.set_option('display.max_rows', None)
    display(df_scoring[['ID_ZONA', 'NOMBRE_ZONA', 'CLUSTER']])
    pd.reset_option('display.max_rows')
else:
    print(f"❌ ERROR: No se encontró el archivo en {archivo_clusters}")

```

```python
# PASO 5: PONDERACIÓN INDIVIDUAL DE 16 VARIABLES (NIVEL MICRO)

print("-" * 70)
print("🚀 PASO 5: PONDERACIÓN INDIVIDUAL (MICRO) - PROTECCIÓN DE DATOS")
print("-" * 70)

# 5.1 Definición de Pesos Locales
pesos = {
    'SCORE_POBLACION_KM2': 0.30, 'SCORE_PORCENTAJE_HISPANOS': 0.20,
    'SCORE_EDAD_MEDIANA': 0.10, 'SCORE_INGRESO_MEDIANO_HOGAR': 0.25,
    'SCORE_TAMANO_HOGAR_PROMEDIO': 0.15,
    'SCORE_MOVILIDAD_PROMEDIO_DIARIA': 0.80, 'SCORE_MOV_CANTIDAD_ESTACIONES': 0.20,
    'SCORE_DELITO_PROPIEDAD_KM2': 0.45, 'SCORE_DELITO_TRANSPORTE_KM2': 0.35,
    'SCORE_DELITO_OTROS_KM2': 0.20,
    'SCORE_LUGARES_COMERCIO_KM2': 0.35, 'SCORE_LUGARES_OFICINAS_KM2': 0.45,
    'SCORE_LUGARES_RESIDENCIAL_KM2': 0.20,
    'SCORE_COMPETENCIA_DIRECTA_KM2': 0.90, 'SCORE_COMPETENCIA_INDIRECTA_KM2': 0.10,
    'SCORE_ALQ_PRECIO_PIE2_ANUAL': 1.00
}

# 5.2 CREACIÓN DE DF_MICRO
df_micro = df_scoring[['ID_ZONA', 'NOMBRE_ZONA', 'CLUSTER']].copy()

for var, peso in pesos.items():
    df_micro[var] = (df_scoring[var] * peso).round(2)

# 5.3 Visualización Completa (31 zonas)
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

print("📋 TABLA MICRO: 16 VARIABLES PONDERADAS INDIVIDUALMENTE")
display(df_micro.sort_values(by=['CLUSTER', 'NOMBRE_ZONA']))

# 5.4 Exportación a resultados/05_scoring
os.makedirs("../../resultados/05_scoring", exist_ok=True)
archivo_micro = "../../resultados/05_scoring/scoring_micro.csv"
df_micro.to_csv(archivo_micro, index=False)

pd.reset_option('display.max_rows')
pd.reset_option('display.max_columns')

print("-" * 70)
print("✅ Nivel Micro protegido. Fuente (0-100) preservada.")
print(f"💾 Archivo MICRO exportado: {archivo_micro}")

```

```python
# PASO 6: CONSOLIDACIÓN POR DIMENSIÓN (NIVEL MACRO)

print("-" * 70)
print("🚀 PASO 6: CONSOLIDACIÓN POR DIMENSIÓN (NIVEL MACRO)")
print("-" * 70)

# 6.1 Creamos un nuevo DataFrame
columnas_control = ['ID_ZONA', 'NOMBRE_ZONA', 'CLUSTER']
df_macro = df_micro[columnas_control].copy()

# 6.2 Consolidación: suma de las subdimensiones ponderadas del Paso 5
df_macro['DEMANDA'] = (
    df_micro['SCORE_POBLACION_KM2'] +
    df_micro['SCORE_PORCENTAJE_HISPANOS'] +
    df_micro['SCORE_EDAD_MEDIANA'] +
    df_micro['SCORE_INGRESO_MEDIANO_HOGAR'] +
    df_micro['SCORE_TAMANO_HOGAR_PROMEDIO']
).round(2)

df_macro['MOVILIDAD'] = (
    df_micro['SCORE_MOVILIDAD_PROMEDIO_DIARIA'] +
    df_micro['SCORE_MOV_CANTIDAD_ESTACIONES']
).round(2)

df_macro['SEGURIDAD'] = (
    df_micro['SCORE_DELITO_PROPIEDAD_KM2'] +
    df_micro['SCORE_DELITO_TRANSPORTE_KM2'] +
    df_micro['SCORE_DELITO_OTROS_KM2']
).round(2)

df_macro['PTOS_INTERES'] = (
    df_micro['SCORE_LUGARES_COMERCIO_KM2'] +
    df_micro['SCORE_LUGARES_OFICINAS_KM2'] +
    df_micro['SCORE_LUGARES_RESIDENCIAL_KM2']
).round(2)

df_macro['COMPETENCIA'] = (
    df_micro['SCORE_COMPETENCIA_DIRECTA_KM2'] +
    df_micro['SCORE_COMPETENCIA_INDIRECTA_KM2']
).round(2)

df_macro['COSTE'] = df_micro['SCORE_ALQ_PRECIO_PIE2_ANUAL'].round(2)

# 6.3 Visualización Total (31 zonas)
pd.set_option('display.max_rows', None)

print("📋 TABLA MACRO: SUMA DE SUBDIMENSIONES PONDERADAS")
print("-" * 60)
display(df_macro.sort_values(by=['CLUSTER', 'NOMBRE_ZONA']))

# 6.4 Exportación
os.makedirs("../../resultados/05_scoring", exist_ok=True)
archivo_macro = "../../resultados/05_scoring/scoring_macro.csv"
df_macro.to_csv(archivo_macro, index=False)

pd.reset_option('display.max_rows')

print("-" * 70)
print(f"✅ Nivel Macro consolidado y exportado: {archivo_macro}")
print("-" * 70)

```
