```python
# PASO 1: ARCHIVO GEOGRAFICO (ZONAS)

import pandas as pd
import geopandas as gpd
import os

# 1. Definición de rutas
ruta_input = "../../datos/limpios"
ruta_sql = "../../datos/base_datos"

# 2. Carga delarchivo geografico y visualización
path_geojson = os.path.join(ruta_input, 'zonas', '12_NTAs_Manhattan_Geometry.geojson')
mapa_raw = gpd.read_file(path_geojson)

# 2.1 Listamos las columnas 
print("\nColumnas detectadas en el GeoJSON:")
print(mapa_raw.columns.tolist())

# 2.2 Visualización de la data
display(mapa_raw.head(10))

# 3. Conversión de Áreas - AREA_M2 a AREA_KM2
mapa_raw['AREA_M2'] = pd.to_numeric(mapa_raw['AREA_M2'], errors='coerce')
mapa_raw['AREA_KM2'] = (mapa_raw['AREA_M2'] / 1_000_000).round(4)


# 4. Seleccionamos las variables a utilziar
mapa_manhattan = mapa_raw[['ID_ZONA', 'NOMBRE_ZONA', 'NOMBRE_DISTRITO', 'AREA_KM2', 'geometry']].copy()


# 5. Exportamos los dataset (tabla y geometria)

# 5.1 Para base_datos 
df_zonas_sql = mapa_manhattan[['ID_ZONA', 'NOMBRE_ZONA', 'NOMBRE_DISTRITO','AREA_KM2']].drop_duplicates()
df_zonas_sql.to_csv(os.path.join(ruta_sql, 'zonas', 'ZONAS.csv'), index=False)

print("\n" + "="*70)
print(f"✅ Tabla ZONAS (CSV) lista en: {os.path.join(ruta_sql, 'zonas')}")
print("="*70)

# 6. Verificación

# 6.1 Verificación del Archivopara tabla de SQL
if os.path.exists(os.path.join(ruta_sql, 'zonas', 'ZONAS.csv')):
    print("\n[OK] ZONAS.csv encontrado en datos/base_datos/zonas")
    df_verif_sql = pd.read_csv(os.path.join(ruta_sql, 'zonas', 'ZONAS.csv'))
    display(df_verif_sql.head(3))
else:
    print("[ERROR] No se encontró ZONAS.csv")

```

```python
# PASO 2: BLOQUE COMPETENCIA

# 1. Carga e inspección
archivo_01 = "01_Restaurantes_Manhattan.csv"
path_01 = os.path.join(ruta_input, 'competencia', archivo_01)
df_01_raw = pd.read_csv(path_01)

print(f"--- 🔍 INSPECCIÓN: {archivo_01} ---")

# 1.1 Visualizamos los datos
display(df_01_raw.head(10))

# 1.2 Verificamos tipos de datos para asegurar que las coordenadas son numéricas
df_01_raw.info()

# 2. Selección de variables y Renombrado inicial
columnas_interes = [
    'ID_RESTAURANTE', 'NOMBRE_RESTAURANTE', 'LATITUD', 'LONGITUD', 
    'COCINA', 'TIPO_COMPETENCIA'
]
df_01_filtro = df_01_raw[columnas_interes].copy()

# Renombramos COCINA a TIPO_COCINA para mayor claridad
df_01_filtro = df_01_filtro.rename(columns={'COCINA': 'TIPO_COCINA'})

# 3. Join Espacial
# Convertimos los puntos en geometría para el cruce
gdf_01 = gpd.GeoDataFrame(
    df_01_filtro, 
    geometry=gpd.points_from_xy(df_01_filtro.LONGITUD, df_01_filtro.LATITUD),
    crs="EPSG:4326"
)

# Cruzamos con 'mapa_manhattan' para obtener el ID_ZONA 
# Eliminamos NOMBRE_ZONA de la selección para evitar redundancia
df_01_unido = gpd.sjoin(
    gdf_01, 
    mapa_manhattan[['ID_ZONA', 'geometry']], 
    how="left", 
    predicate="within"
)

# Limpieza: eliminamos columnas espaciales y de sistema para el CSV final del proyecto
df_01_proyecto = df_01_unido.drop(columns=['index_right', 'geometry']).copy()

# Verificación del Join
print(f"\nRestaurantes sin barrio asignado: {df_01_proyecto['ID_ZONA'].isna().sum()}")

# 4. Exportamos la tabla para base_datos
filas_p, cols_p = df_01_proyecto.shape

# 5. Agrupación para SQL: Resumen por Barrio y Tipo de Cocina
# Agrupamos solo por ID_ZONA (la llave) y TIPO_COCINA
df_sql_01 = df_01_proyecto.groupby(['ID_ZONA', 'TIPO_COCINA']).agg(
    CLASIFICACION_COMPETENCIA=('TIPO_COMPETENCIA', 'first'),
    CANTIDAD_LOCALES=('ID_RESTAURANTE', 'count')
).reset_index()

# Guardamos el archivo final para la base de datos SQL
df_sql_01.to_csv(os.path.join(ruta_sql, 'competencia', 'RESTAURANTES.csv'), index=False)
filas_s, cols_s = df_sql_01.shape

print("\n" + "="*50)
print("✅ PASO 2 COMPLETADO: Tabla Restaurantes por Cocina")
print(f"📊 Registros procesados: {filas_p} locales individuales.")
print(f"📊 SQL: {filas_s} categorías de competencia por zona exportadas.")
print("="*50)

# Verificación de la tabla para SQL
display(df_sql_01.head(15))
```

```python
# PASO 3: BLOQUE SEGURIDAD

# 1. Carga e inspección
archivo_02 = "02_Seguridad_Manhattan.csv"
path_02 = os.path.join(ruta_input, 'seguridad', archivo_02)
df_02_raw = pd.read_csv(path_02)

print(f"--- 🔍 INSPECCIÓN: {archivo_02} ---")

# 1.1 Visualizamos los datos
display(df_02_raw.head(10))

# 1.2 Verificamos tipos de datos
df_02_raw.info()

# 2. Transformación de delitos y lugares

# 2.1 Mapeo de delitos a TIPO_DELITO
dicc_delitos = {
    'FELONY ASSAULT': 'ASALTO GRAVE',
    'ROBBERY': 'ROBO CON VIOLENCIA',
    'BURGLARY': 'ROBO A ESTABLECIMIENTO'
}
df_02_raw['TIPO_DELITO'] = df_02_raw['DELITO'].map(dicc_delitos)

# 2.2 Categorización de lugares a LUGAR_DELITO
def categorizar_lugar(lugar):
    l = str(lugar).upper()
    if any(x in l for x in ['TRANSIT', 'SUBWAY', 'BUS', 'TAXI', 'FERRY', 'TERMINAL']):
        return 'TRANSPORTE'
    elif any(x in l for x in ['STREET', 'PARK', 'OPEN AREAS', 'HIGHWAY', 'BRIDGE', 'SIDEWALK']):
        return 'VIA PUBLICA'
    elif any(x in l for x in ['STORE', 'COMMERCIAL', 'RESTAURANT', 'SHOP', 'BANK', 'HOTEL', 'BAR', 'GYM']):
        return 'LOCAL COMERCIAL'
    elif any(x in l for x in ['RESIDENCE', 'HOUSE', 'APARTMENT']):
        return 'ZONA RESIDENCIAL'
    elif any(x in l for x in ['HOSPITAL', 'CHURCH', 'SCHOOL', 'BUILDING', 'UNIVERSITY', 'OFFICE']):
        return 'EDIFICIOS PUBLICOS'
    else:
        return 'OTROS'

df_02_raw['LUGAR_DELITO'] = df_02_raw['TIPO_LUGAR'].apply(categorizar_lugar)


# 3. Selección y Join Espacial

# Seleccionamos las columnas con los nombres definitivos
columnas_interes = ['ID_QUEJA', 'LATITUD', 'LONGITUD', 'TIPO_DELITO', 'LUGAR_DELITO']
df_02_filtro = df_02_raw[columnas_interes].copy()

gdf_02 = gpd.GeoDataFrame(
    df_02_filtro, 
    geometry=gpd.points_from_xy(df_02_filtro.LONGITUD, df_02_filtro.LATITUD),
    crs="EPSG:4326"
)

# Cruzamos con 'mapa_manhattan' para obtener solo ID_ZONA (Sin Nombre Zona para evitar redundancia)
df_02_unido = gpd.sjoin(
    gdf_02, 
    mapa_manhattan[['ID_ZONA', 'geometry']], 
    how="left", 
    predicate="within"
)

# Limpieza de columnas técnicas
df_02_proyecto = df_02_unido.drop(columns=['index_right', 'geometry']).copy()

# Verificación del Join
print(f"\nIncidentes sin barrio asignado: {df_02_proyecto['ID_ZONA'].isna().sum()}")

# 4. Exportamos los datasets

# Registros procesados para integración
filas_p, cols_p = df_02_proyecto.shape

# SQL: Tabla resumen normalizada (ID_ZONA + Tipos + Conteo)
df_sql_02 = df_02_proyecto.groupby(['ID_ZONA', 'TIPO_DELITO', 'LUGAR_DELITO']).agg(
    CANTIDAD_INCIDENTES=('ID_QUEJA', 'count')
).reset_index()

# Guardamos la tabla para SQL
df_sql_02.to_csv(os.path.join(ruta_sql, 'seguridad', 'SEGURIDAD.csv'), index=False)
filas_s, cols_s = df_sql_02.shape

print("\n" + "="*50)
print("✅ PASO 3 COMPLETADO: Tabla Seguridad Normalizada")
print(f"📊 Registros procesados: {filas_p} incidentes individuales.")
print(f"📊 SQL: {filas_s} filas con conteos por tipo y lugar.")
print("="*50)

# Verificación de la tabla final
display(df_sql_02.head(15))
```

```python
# PASO 4: BLOQUE MOVILIDAD (MTA)

# 1. Carga e inspección
archivo_10 = "10_MTA_Manhattan.csv"
path_10 = os.path.join(ruta_input, 'movilidad', archivo_10)
df_10_raw = pd.read_csv(path_10)

print(f"--- 🔍 INSPECCIÓN: {archivo_10} ---")

# 1.1 Visualizamos los datos 
display(df_10_raw.head(10))

# 1.2 Verificamos tipos de datos
df_10_raw.info()

# 2. PROCESAMIENTO: Agregación por Estación y Tipo de Día
# Agregamos por día (Sumamos los pasajeros que pasaron entre 10am y 10pm)
df_diario = df_10_raw.groupby(['ID_ESTACION', 'ESTACION', 'LATITUD', 'LONGITUD', 'FECHA', 'TIPO_DIA']).agg(
    TOTAL_PASAJEROS_DIA=('PASAJEROS', 'sum')
).reset_index()

# Calculamos el PROMEDIO DIARIO (Laborable vs Fin de Semana)
df_mta_promedio = df_diario.groupby(['ID_ESTACION', 'ESTACION', 'LATITUD', 'LONGITUD', 'TIPO_DIA']).agg(
    PROMEDIO_DIARIO_PASAJEROS=('TOTAL_PASAJEROS_DIA', 'mean')
).reset_index()

# Convertimos a ENTERO para que sea un número de personas real
df_mta_promedio['PROMEDIO_DIARIO_PASAJEROS'] = df_mta_promedio['PROMEDIO_DIARIO_PASAJEROS'].round(0).astype(int)

# 3. Join Espacial (Asignar ID_ZONA)
gdf_mta = gpd.GeoDataFrame(
    df_mta_promedio, 
    geometry=gpd.points_from_xy(df_mta_promedio.LONGITUD, df_mta_promedio.LATITUD),
    crs="EPSG:4326"
)

# Cruzamos con 'mapa_manhattan' para obtener solo ID_ZONA (Sin Nombre de Zona por normalización)
df_mta_unido = gpd.sjoin(
    gdf_mta, 
    mapa_manhattan[['ID_ZONA', 'geometry']], 
    how="left", 
    predicate="within"
)

# Si quedan nulos, usamos 'sjoin_nearest' para que busquen el barrio más cercano 
if df_mta_unido['ID_ZONA'].isna().any():
    # Separamos los nulos
    nulos = df_mta_unido[df_mta_unido['ID_ZONA'].isna()].copy()
    nulos = nulos.drop(columns=['ID_ZONA', 'index_right'])
    
    # PROYECTAMOS A METROS (EPSG:2263 es para NY) para que el cálculo sea exacto
    nulos_proy = nulos.to_crs("EPSG:2263")
    mapa_proy = mapa_manhattan.to_crs("EPSG:2263")
    
    # Buscamos el vecino más cercano
    nulos_corregidos = gpd.sjoin_nearest(
        nulos_proy, 
        mapa_proy[['ID_ZONA', 'geometry']], 
        how="left"
    )
    
    # Volvemos a unir los datos corregidos al dataframe principal
    df_mta_unido = pd.concat([
        df_mta_unido[df_mta_unido['ID_ZONA'].notna()], 
        nulos_corregidos.to_crs("EPSG:4326")
    ])

# Limpieza final del Join
df_mta_final = df_mta_unido.drop(columns=['index_right', 'geometry']).copy()

# 4. Exportamos los datasets

# Registros procesados para integración
filas_p, cols_p = df_mta_final.shape

# SQL: Resumen por Barrio y Tipo de Día (Normalizada: Solo ID_ZONA)
df_sql_10 = df_mta_final.groupby(['ID_ZONA', 'TIPO_DIA']).agg(
    VOLUMEN_TOTAL_PASAJEROS=('PROMEDIO_DIARIO_PASAJEROS', 'sum'),
    CANTIDAD_ESTACIONES=('ID_ESTACION', 'nunique')
).reset_index()

# Aseguramos que el resultado sea un número entero limpio
df_sql_10['VOLUMEN_TOTAL_PASAJEROS'] = df_sql_10['VOLUMEN_TOTAL_PASAJEROS'].astype(int)

# Guardamos la tabla definitiva para la base de datos
df_sql_10.to_csv(os.path.join(ruta_sql, 'movilidad', 'MOVILIDAD.csv'), index=False)
filas_s, cols_s = df_sql_10.shape

print("\n" + "="*60)
print("✅ MOVILIDAD ACTUALIZADA: Tabla normalizada para SQL")
print(f"📊 Registros procesados: {filas_p} registros de estaciones.")
print(f"📊 SQL: {filas_s} filas con flujos agregados por zona.")
print("="*60)

# Verificación: Barrios con mayor flujo de personas
display(df_sql_10.sort_values(by='VOLUMEN_TOTAL_PASAJEROS', ascending=False).head(10))
```

```python
# PASO 5: BLOQUE LUGARES DE INTERÉS (11 - Common Places)

# 1. Carga e inspección
archivo_11 = "11_Common_Places_Manhattan.csv"
path_11 = os.path.join(ruta_input, 'lugares_interes', archivo_11)
df_11_raw = pd.read_csv(path_11)

print(f"--- 🔍 INSPECCIÓN: {archivo_11} ---")

# 1.1 Visualizamos los datos 
display(df_11_raw.head(10))

# 1.2 Verificamos tipos de datos
df_11_raw.info()

# 1.3 Revisamos las categorías únicas y renombramos la columna
print("\nCategorías detectadas en Lugares Comunes:")
print(df_11_raw['CATEGORIA'].unique())

# Renombramos CATEGORIA a CATEGORIA_LUGAR para mayor claridad técnica
df_11_raw = df_11_raw.rename(columns={'CATEGORIA': 'CATEGORIA_LUGAR'})

# 2. Selección de variables
columnas_interes = ['NOMBRE_LUGAR', 'CATEGORIA_LUGAR', 'LATITUD', 'LONGITUD']
df_11_filtro = df_11_raw[columnas_interes].copy()

# 3. Join Espacial
# Convertimos a geometría
gdf_11 = gpd.GeoDataFrame(
    df_11_filtro, 
    geometry=gpd.points_from_xy(df_11_filtro.LONGITUD, df_11_filtro.LATITUD),
    crs="EPSG:4326"
)

# Cruzamos con 'mapa_manhattan' para asignar solo el ID_ZONA (Normalización)
df_11_unido = gpd.sjoin(
    gdf_11, 
    mapa_manhattan[['ID_ZONA', 'geometry']], 
    how="left", 
    predicate="within"
)

# 3.1 Corrección de puntos fuera de límites (Sjoin Nearest)
if df_11_unido['ID_ZONA'].isna().any():
    nulos = df_11_unido[df_11_unido['ID_ZONA'].isna()].copy()
    nulos = nulos.drop(columns=['ID_ZONA', 'index_right'])
    
    # Proyectamos a metros para precisión en el borde de la isla (EPSG:2263 NY)
    nulos_proy = nulos.to_crs("EPSG:2263")
    mapa_proy = mapa_manhattan.to_crs("EPSG:2263")
    
    nulos_corregidos = gpd.sjoin_nearest(
        nulos_proy, 
        mapa_proy[['ID_ZONA', 'geometry']], 
        how="left"
    )
    
    # Combinamos resultados
    df_11_unido = pd.concat([
        df_11_unido[df_11_unido['ID_ZONA'].notna()], 
        nulos_corregidos.to_crs("EPSG:4326")
    ])

# Limpieza de columnas técnicas del Join
df_11_proyecto = df_11_unido.drop(columns=['index_right', 'geometry']).copy()

# Verificación final
print(f"✅ Lugares sin barrio tras la corrección: {df_11_proyecto['ID_ZONA'].isna().sum()}")
display(df_11_proyecto.head(5))

# 4. Exportamos los datasets

filas_p, cols_p = df_11_proyecto.shape

# 5. Tabla para SQL (Normalizada: ID_ZONA + CATEGORIA_LUGAR)
df_sql_11 = df_11_proyecto.groupby(['ID_ZONA', 'CATEGORIA_LUGAR']).agg(
    CANTIDAD_LUGARES=('NOMBRE_LUGAR', 'count')
).reset_index()

# Guardamos la tabla para SQL
df_sql_11.to_csv(os.path.join(ruta_sql, 'lugares_interes', 'LUGARES_INTERES.csv'), index=False)
filas_s, cols_s = df_sql_11.shape

print("\n" + "="*50)
print("✅ PASO 5 COMPLETADO: Tabla Lugares Normalizada")
print(f"📊 Registros procesados: {filas_p} lugares individuales.")
print(f"📊 SQL: {filas_s} categorías por zona exportadas.")
print("="*50)

# Verificación de la tabla para SQL
display(df_sql_11.head(15))
```

```python
# PASO 6.1: BLOQUE CENSO - POBLACIÓN 

pd.set_option('display.max_columns', None)

# 1. Carga e inspección
archivo_03 = "03_Censo_Age_Sex_Manhattan.csv"
path_03 = os.path.join(ruta_input, 'censo', archivo_03)
df_03_raw = pd.read_csv(path_03)

# Visualizamos para identificar los nombres exactos de las columnas de población y edad
display(df_03_raw.head(5))
df_03_raw.info()

# 2. Selección de variables y limpieza de nombres
df_03_filtro = df_03_raw[['Zona_Codigo', 'Total Poblacion', 'Media de Edad']].copy()

# Renombramos para estandarizar con base maestra
df_03_filtro = df_03_filtro.rename(columns={
    'Zona_Codigo': 'ID_ZONA',
    'Total Poblacion': 'POBLACION_TOTAL',
    'Media de Edad': 'EDAD_MEDIANA'
})

# 5. Métricas de exportación
filas, cols = df_03_filtro.shape

# 7. Verificación de Calidad
print("\n" + "="*70)
print(f" VERIFICACION DE LOS DATOS")
print(f"✅ Variable POBLACION_TOTAL: Rango [{df_03_filtro['POBLACION_TOTAL'].min()} - {df_03_filtro['POBLACION_TOTAL'].max()}]")
print(f"✅ Variable EDAD_MEDIANA: Promedio general de Manhattan: {df_03_filtro['EDAD_MEDIANA'].mean():.2f}")

print("\n" + "="*70)
print(f"✅ PASO 6.1 COMPLETADO: {archivo_03}")
print(f"📊 Variables: ID_ZONA, POBLACION_TOTAL, EDAD_MEDIANA")
print(f"📊 Dataset: {filas} filas y {cols} columnas.")
print("="*70)

display(df_03_filtro.head(31))
```

```python
# PASO 6.2: BLOQUE CENSO - ORIGEN HISPANO

pd.set_option('display.max_columns', None)

# 1. Carga e inspección
archivo_04 = "04_Censo_Hispanic_Origin_Manhattan.csv"
path_04 = os.path.join(ruta_input, 'censo', archivo_04)
df_04_raw = pd.read_csv(path_04)

#visualizamos
display(df_04_raw.head(5))

# 2. AGRUPACIÓN: Sumamos la población por 'Grupo'
df04_agrupado = df_04_raw.groupby(['Zona_Nombre', 'Zona_Codigo', 'Grupo'])['Poblacion'].sum().reset_index()

# 3. PIVOTAJE: Convertimos los grupos en columnas
df04_pivot = df04_agrupado.pivot(
    index=['Zona_Nombre', 'Zona_Codigo'], 
    columns='Grupo', 
    values='Poblacion'
).reset_index()

# 4. RENOMBRADO Y SELECCIÓN: Estandarizamos nombres
df04_refinado = df04_pivot.rename(columns={
    'Zona_Codigo': 'ID_ZONA',
    'Hispanic or Latino': 'POBLACION_HISPANA'
})

# 5. UNIÓN A LA BASE MAESTRA (union con la data anterior de poblacion)
    
df_censo_consolidado = pd.merge(
    df_03_filtro, 
    df04_refinado[['ID_ZONA', 'POBLACION_HISPANA']], 
    on='ID_ZONA', 
    how='left'
)

# 6. Verificación de Calidad
filas, cols = df_censo_consolidado.shape
print("\n" + "="*70)
print(f"✅ PASO 6.2 COMPLETADO: Unión de {archivo_04}")
print(f"📊 Variables actuales: {df_censo_consolidado.columns.tolist()}")
print(f"📊 Dataset actual: {filas} filas y {cols} columnas.")
print(f"🏠 Poblacion Hispana: [{df_censo_consolidado['POBLACION_HISPANA'].min()} - {df_censo_consolidado['POBLACION_HISPANA'].max()}]")

print("="*70)

display(df_censo_consolidado.head(31))
```

```python
# PASO 6.4: BLOQUE CENSO - TASA DE EMPLEO

# 1. Carga e inspección
archivo_06 = "06_Censo_Employment_Status_Manhattan.csv"
path_06 = os.path.join(ruta_input, 'censo', archivo_06)
df_06_raw = pd.read_csv(path_06)

#Visualizamos
display(df_06_raw.head(5))

# 2. CÁLCULO DE LA TASA: (Employed / Pöpulation 16 years and over * 100
df_06_raw['TASA_EMPLEO'] = (
    df_06_raw['Employed'] / df_06_raw['Population 16 years and over']
) * 100

df_06_raw['TASA_EMPLEO'] = df_06_raw['TASA_EMPLEO'].round(2)

# 3. Selección y renombre
df_06_refinado = df_06_raw[['Zona_Codigo', 'TASA_EMPLEO']].copy()
df_06_refinado = df_06_refinado.rename(columns={'Zona_Codigo': 'ID_ZONA'})

# 4. UNIÓN AL CONSOLIDADO 
# Antes, borramos si existe por error previo
if 'TASA_EMPLEO' in df_censo_consolidado.columns:
    df_censo_consolidado = df_censo_consolidado.drop(columns=['TASA_EMPLEO'])

df_censo_consolidado = pd.merge(
    df_censo_consolidado, 
    df_06_refinado[['ID_ZONA', 'TASA_EMPLEO']], 
    on='ID_ZONA', 
    how='left'
)

# 4. Verificación de Calidad
filas, cols = df_censo_consolidado.shape
print("\n" + "="*70)
print(f"✅ PASO 6.4 COMPLETADO: {archivo_06}")
print(f"💼 Promedio de empleo en las zonas: {df_censo_consolidado['TASA_EMPLEO'].mean():.2f}%")
print(f"📊 Dataset actual: {filas} filas y {cols} columnas.")
print(f"💼 Rango detectado: [{df_censo_consolidado['TASA_EMPLEO'].min():.2f}% - {df_censo_consolidado['TASA_EMPLEO'].max():.2f}%]")
print("="*70)

display(df_censo_consolidado.head(31))
```

```python
# PASO 6.6: BLOQUE CENSO - OCUPACIÓN DE VIVIENDA

# 1. Carga e inspección
archivo_08 = "08_Censo_Housing_Occupancy_Manhattan.csv"
path_08 = os.path.join(ruta_input, 'censo', archivo_08)
df_08_raw = pd.read_csv(path_08)

#Visualizamos
display(df_08_raw.head(5))

# 2. CÁLCULO DE LA TASA DE OCUPACIÓN
# (Unidades Ocupadas / Unidades Totales) * 100
df_08_raw['TASA_OCUPACION'] = (
    df_08_raw['Occupied housing units'] / df_08_raw['Total housing units']
) * 100

# Redondeamos a 2 decimales como las variables anteriores
df_08_raw['TASA_OCUPACION'] = df_08_raw['TASA_OCUPACION'].round(2)

# 3. Seleccion
df_08_refinado = df_08_raw[['Zona_Codigo', 'TASA_OCUPACION']].rename(columns={'Zona_Codigo': 'ID_ZONA'})

# 4. UNIÓN AL CONSOLIDADO
# Limpiamos si ya existía por error
if 'TASA_OCUPACION' in df_censo_consolidado.columns:
    df_censo_consolidado = df_censo_consolidado.drop(columns=['TASA_OCUPACION'])

df_censo_consolidado = pd.merge(
    df_censo_consolidado, 
    df_08_refinado, 
    on='ID_ZONA', 
    how='left'
)

# 5. Verificación de Calidad
filas, cols = df_censo_consolidado.shape
print(f"✅ PASO 6.6 COMPLETADO: {archivo_08}")
print(f"📊 Variables actuales: {df_censo_consolidado.columns.tolist()}")
print(f"📊 Dataset actual: {filas} filas y {cols} columnas.")
print(f"🏠 Ocupación promedio en las zonas: [{df_censo_consolidado['TASA_OCUPACION'].min()} - {df_censo_consolidado['TASA_OCUPACION'].max()}]")
print("="*70)

display(df_censo_consolidado.head(31))
```
